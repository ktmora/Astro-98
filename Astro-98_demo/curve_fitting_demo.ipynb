```python
# Curve fitting demo :
```

```python
import matplotlib.pyplot as plt
import scipy.optimize as fit
```

```python
m = 0.5
b = 3
sigma = 1
n = 20
xdata = np.arange(0, n)
ydata = m * xdata + b + sigma * np.random.standard_normal(size = n)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(xdata, ydata)
plt.show()
```

```python
def f(x, m, b):
    return m*x + b

p0 = [0.3, 1]

parameters, covariance_matrix = fit.curve_fit(f, xdata, ydata, p0)
```

```python
parameters
```

```python
covariance_matrix
```

```python
m_error = covariance_matrix [0,0] ** 0.5
b_error = covariance_matrix[1,1] ** 0.5
```

```python
m = parameters[0]
b = parameters[1]
```

```python
print('m={0:1.5f} +\- {1:1.5f}'.format(m,m_error))

# m = 0.48224 +\- 0.03446
```

```python
x = np.linspace(xdata.min(), xdata.max(), 100)
y = f(x, m ,b)
plt.figure()
plt.plot(x,y)
plt.scatter(xdata,ydata)
plt.show()
```
