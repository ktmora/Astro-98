```python
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
%matplotlib inline
```

```python
x, y = np.meshgrid(np.linspace(-1, 1, 100), np.linspace(-2, 2, 100)) #np.linspace(range, range, step)

z = x + y # this is my 'f' function normally, on the z axis

plt.figure()
plt.imshow(z, cmap = 'inferno' )
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.colorbar()
plt.show()
```

```python
x, y = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100)) 
z = -(np.sqrt(x**2 + y**2))

plt.figure()
plt.imshow(z, cmap = 'inferno' )
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.colorbar()
plt.show()
```
