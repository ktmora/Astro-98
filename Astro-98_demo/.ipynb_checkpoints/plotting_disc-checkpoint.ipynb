```python
#Plotting Discussion question 1 
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,3,100)
yerr = 0.05 + 2*x
```

```python
plt.figure()
for p in range(2,6):
    #plt.plot(x,x**p, label = 'x^' + str(p) )
    plt.errorbar(x,x**p, yerr = yerr, label = 'x^' + str(p))

#plt.plot(x,x**3, label = 'x^3')
#plt.plot(x,x**4, label = 'x^4')
#plt.plot(x,x**5, label = 'x^5')

plt.title("Exponential Plots")
plt.xlabel('X Values')
plt.ylabel('Exponents')
plt.legend()
plt.show()
```

```python
#question 3
image_data = np.loadtxt('astro_image.csv')
```

```python
np.shape(image_data)
```

```python
plt.figure()
plt.imshow(image_data)
plt.show()
```

```python
#Things we can do to make this better: Change the Colormap
plt.figure()
plt.imshow(image_data, cmap = 'inferno', vmin = 50, vmax = 400) 
plt.colorbar()
plt.show()
```

```python
#way better than what we started with!
```

```python
#can also change where the origin is, and sometimes it is better as upper or lower for diff data.
#Things we can do to make this better: Change the Colormap
plt.figure()
plt.imshow(image_data, cmap = 'inferno', vmin = 50, vmax = 400, origin = 'lower') 
#lowering the vmax does not remove the data, it just reshapes the data and fits it to this scale. so basically re-scaling it
#to make it more purdy
#A lot of finding the vmax is just messing around with the values until it looks right.
plt.colorbar()
plt.show()
```

```python
#clearly this data should have the origin at the top.
```

```python
#question 2
x = np.linspace(0,5,100)

y1 = np.exp(x)
y2 = np.exp(2*x)
y3 = np.exp(x)+1
y4 = np.exp(x)-1

fig,axs = plt.subplots(2,2)

axs[0,0].plot(x,y1)
axs[0,1].plot(x,y2)
axs[1,0].plot(x,y3)
axs[1,1].plot(x,y4)
plt.tight_layout()
plt.show()
```

```python
#can also set labels and titles for each subplot
#look up code line that allows you to fix the layout : plt.tight_layout()
```
