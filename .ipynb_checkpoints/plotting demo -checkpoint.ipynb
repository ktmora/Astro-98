```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,100)
y = np.sin(x)
```

```python
plt.figure()
plt.plot(x,y)
plt.title('sine')
plt.xlabel('x')
plt.ylabel('y') #all the stuff inside the () for plot labels has to be a string
plt.show()
```

```python
x2 = np.linspace(0, 2*np.pi, 100)
y2 = x2**2
```

```python
plt.figure(figsize = (7.3, 4))
plt.plot(x2, y2, color = 'thistle', marker = '2', linestyle = 'dashdot')
plt.title('x squared')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
