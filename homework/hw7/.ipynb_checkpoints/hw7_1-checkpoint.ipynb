```python
import numpy as np
import pandas as pd
```

```python
#Load in text file AOC_recent_tweets.txt

df_tweets = pd.read_json('AOC_recent_tweets.txt')
df_tweets
```

```python
#Index by the 'id' col.
df_tweets.iloc[:, 1:3:2]
```

```python
#Create a function that converts the creation time and to a time as an hour in decimal form. 
list_dec_hour = []
def time_in_hours( ):
    created_at = df_tweets.loc[:,'created_at']
    for i in range(len(created_at)) : 
        time = pd.to_datetime(created_at.loc[i])
        dec_hour = time.hour + (time.minute/60) + (time.second/60)/60
        list_dec_hour.append(dec_hour)
    return (list_dec_hour)
print(time_in_hours( ))
```

```python
#Adds a column at the end of df_tweets called 'hours'

hour = pd.Series(list_dec_hour)
df_tweets['hours'] = hour.values
df_tweets['hours']
df_tweets.insert(1, 'hours', [hour.values])
#display(df_tweets)
```

```python
#Index + shows the new column added, 'hours'. 
modified_tweets_df = df_tweets.loc[:, ['created_at', 'hours', 'full_text']]
display(modified_tweets_df)
```

```python
#Creates a .csv file from the above pandas data frame.
modified_tweets_df.to_csv('modified_tweets2.csv', index = False)
```
