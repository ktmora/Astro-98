```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Loading + displaying the data.

planets_df = sns.load_dataset('planets')
planets_df
```

```python
#Plot the orbital period (x-axis) vs the mass (y-axis) of planets from the dataset 'planets'.

plt.figure()

sns.scatterplot(data=planets_df, x = 'orbital_period', y = 'mass', hue = 'method')

#Using log scale to make the data more readable.
plt.xscale('log')
plt.yscale('log')

#Axis Labels
plt.title('Orbital Period vs Mass of Planets')
plt.xlabel('Log Orbital Period')
plt.ylabel('Log Mass')

#Legend off plot
ax = sns.scatterplot(data=planets_df, x = 'orbital_period', y = 'mass', hue = 'method')
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))

plt.show()
```

```python
#Bar plot displaying the number of exoplanets discovered per year, and color depends on discovery method.

plt.figure(figsize = (10, 6))

sns.histplot(data = planets_df, x = 'year', multiple = 'stack', hue = 'method')

#Got the two lines below from seaborn website : https://seaborn.pydata.org/generated/seaborn.move_legend.html .
ax = sns.histplot(data = planets_df, x = 'year', multiple = 'stack', hue = 'method')
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
#Moves the legend outside of the plot.

#Axis Labels
plt.title('Number of Exoplanets Discovered per Year')
plt.xlabel('Year')
plt.ylabel('Number of Planets Discovered')

plt.show()
```
