```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#part one of question one: make three lists each 40 elements long with values ranging from 0-11
m = 0.5
b = 3
sigma = 1
n = 20
xdata = np.arange(0, n)
ydata = m * xdata + b + sigma * np.random.standard_normal(size = n)
```

```python
rng1 = np.random.default_rng()
rng1 = rng1.integers(low = 0, high = 100, size = 40)
print(rng1)
```

```python
rng2 = np.random.default_rng()
rng2 = rng2.integers(low = 0, high = 100, size = 40)
print(rng2)
```

```python
rng3 = np.random.default_rng()
rng3 = rng3.integers(low = 0, high = 100, size = 40)
print(rng3)
```

```python
# Got the code to create a random array (rng) from : 
# https://numpy.org/doc/stable/reference/random/index.html#module-numpy.random
```

```python
#part 2 of question one: make two subplots.

plt.figure()
fig, axs = plt.subplots(2)
fig.suptitle('Stacked Subplots')

xaxisone = np.arange(40)
yaxisone = rng1
axs[0].plot(xaxisone, yaxisone, color= 'sandybrown', linewidth= 10)
#sandybrown is a shade of orange. I just didn't like the default orange.

yaxis = rng2
axs[1].plot(xaxis, yaxis, color='red', linestyle='dotted')

plt.show()
```

```python
#part 3 of question none: final plot

plt.figure()
xaxis = np.arange(40)
yaxis = rng3
plt.scatter(xaxis, yaxis, color='magenta' , marker= 'D')
plt.show()
```
