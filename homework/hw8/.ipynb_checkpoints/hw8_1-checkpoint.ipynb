```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
from datetime import datetime
```

```python
#Load in the data as a csv
global_land_temp_by_state_df = pd.read_csv('GlobalLandTemperaturesByState.csv')
global_land_temp_by_state_df
```

```python
#Filter the data to only include the date, average termpatrue, and state after row 2000.
data = global_land_temp_by_state_df[global_land_temp_by_state_df['dt'] > '2000']
data = data.loc[:, ['dt','AverageTemperature','State']]
data
```

```python
#Only show Nebraska, Wyoming, and South Dakota in the 'State' column.
mydata = data[(data['State'] == 'Wyoming') | (data['State'] == 'Nebraska') | (data['State'] == 'South Dakota')]
mydata
```

```python
#Modify so that the average temperature for each date for the three states shows.
#There should be two columns: one for the date and on for the avg. temp.
```

```python
i = 0
date = []
avg_temp =[]
for i in range(len(mydata)):
    test_data = mydata[(mydata['dt'] == mydata.iloc[i, 0])] # groups the data by date
    average_temp_per_day = (sum(test_data.iloc[:, 1])) / 3 #finds the avg temp accross the 3 states every date
    avg_temp.append(average_temp_per_day)
    i += 1
# print(avg_temp)
for i in range(len(mydata)):
    date.append(mydata.iloc[i,0])
    i += 1
# print(len(date))
# print(len(avg_temp))
d = {"date":date, "average temperature":avg_temp}
data_df = pd.DataFrame(d)
data_df
```

```python
#Converting the date column to a number that can be plotted
data_df['date'] = pd.to_datetime(data_df['date'])
data_df['new date'] = data_df['date'].dt.year + (data_df['date'].dt.day * (0.01)) + (data_df['date'].dt.month * (0.0001))
data_df.loc[:, ['new date', 'average temperature']]
```

```python
sorted_data = data_df.sort_values(by = 'date')
sorted_data = sorted_data.reset_index(drop = True)
sorted_data.loc[:, ['date', 'average temperature']]
```

```python
plt.figure(figsize = (10,6))
#Plotting the data
xaxis = sorted_data.loc[:, 'date']
yaxis = sorted_data.loc[:, 'average temperature']
plt.plot(xaxis, yaxis)

#Plotting the initial guess of the curve 
x = np.arange(10950, 16000, 5050/495)
T = 1.988/np.pi
s = 0.8
A = 16
k = 8.23
y = np.sin(T*x + s)*A + k
plt.plot(x, y)
plt.show()
```

```python
import scipy.optimize as fit
```

```python
def f(x, T, s, A, k):
    return np.sin(T*x + s)*A + k

p0 = [(1.988/np.pi), 0.8, 16, 8.23]

parameters, covariance_matrix = fit.curve_fit(f, x, y, p0)
```
