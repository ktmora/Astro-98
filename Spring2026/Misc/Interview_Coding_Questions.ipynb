```python
# 2 : Arrays
# Modify this function such that if each element in one row is the equivalent, it returns "True". 
# Note: np.all will return "True" if ALL values passed through are true for a given parameter. 

import numpy as np

arr = np.array([[1, 1, 1],
                [4, 5, 6],
                [2, 2, 2]])

def find_equal(row):
    return np.all(row == row[0])

for row in arr:
    print(find_equal(row))
```

```python
# 3 : While Loop
# Write a while loop that counts backwards from 20 in intervals of 2, and includes 0.  

i = 20
while i > 0:
    i -= 2
    print(i)

# Now, write a function including your previous while loop that returns a numpy array of the values from the loop.

import numpy as np

def i_can_count(arr):
    i = 20 
    arr = np.array(20)
    #insert while loop here
    while i > 0:
        i -= 2
        arr = np.append(arr, i)
    print(arr)

i_can_count(arr)
```
