```python
#Python_DeCal_Logo

#testing code from AI to see what it prints
import turtle
# Set up the screen
screen = turtle.Screen()
screen.bgcolor("black")
screen.title("Logo - Yasin")
# Create the turtle object
pen = turtle.Turtle()
pen.speed(10)
pen.pensize(3)
# Function to draw a circle with text inside
def draw_circle_with_text(x, y, radius, text, color):
   pen.penup()
   pen.goto(x, y - radius)
   pen.pendown()
   pen.fillcolor(color)
   pen.begin_fill()
   pen.circle(radius)
   pen.end_fill()
   # Write text inside the circle
   pen.penup()
   pen.goto(x, y - radius / 2)
   pen.color("white")
   pen.write(text, align="center", font=("Arial", 16, "bold"))
# Draw the logo
draw_circle_with_text(0, 100, 100, "Yasin", "#306998") # Blue circle with text
pen.penup()
pen.goto(-50, -50)
pen.pendown()
# Add additional design (optional)
pen.color("#FFD43B") # Yellow color
pen.begin_fill()
for _ in range(4):
   pen.forward(100)
   pen.right(90)
pen.end_fill()
# Hide the turtle and display the result
pen.hideturtle()
screen.mainloop()
```
